```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv("/content/data.csv",encoding="ISO=8859-1")
heart_df = pd.read_csv("/content/heart.csv")
```

```python
df.info()
```

a. Data Cleaning

```python
df.isnull().sum()
df.head()
```

```python
dfClean = df.drop(columns=['pm2_5','location_monitoring_station','stn_code'])
```

```python
#numerical features
dfClean['so2'] = dfClean['so2'].fillna(dfClean['so2'].median())
dfClean['no2'] = dfClean['no2'].fillna(dfClean['no2'].median())
dfClean['rspm'] = dfClean['rspm'].fillna(dfClean['rspm'].median())
dfClean['spm'] = dfClean['spm'].fillna(dfClean['spm'].median())
```

```python
dfClean['sampling_date'] = dfClean['sampling_date'].fillna(dfClean['sampling_date'].mode()[0])
dfClean['agency'] = dfClean['agency'].fillna(dfClean['agency'].mode()[0])
dfClean['type'] = dfClean['type'].fillna(dfClean['type'].mode()[0])
```

b. data integration

```python
integrated_df = pd.concat([dfClean , heart_df ])
```

```python
integrated_df.info()
```

c. Data Transformation

```python
numericalCols = dfClean[['so2','no2','spm','rspm']]
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
numericalCols.isnull().sum()
```

```python
df2 = dfClean.copy()
df2['so2'] = scaler.fit_transform(df2[['so2']])
df2['no2'] = scaler.fit_transform(df2[['no2']])
df2['spm'] = scaler.fit_transform(df2[['spm']])
df2['rspm'] = scaler.fit_transform(df2[['rspm']])
```

```python
df2.describe()
```

d. Error Correcting

```python
dfClean.duplicated().sum()
#drop duplicates
dfClean = dfClean.drop_duplicates()
```

e. model buliding

```python
from sklearn.model_selection import train_test_split as tts
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error as mse
```

```python
heart_df.tail()
```

```python
X = heart_df[['age','chol','trestbps','thalach']]
Y = heart_df[['target']]
```

```python
x_train , x_test , y_train , y_test = tts(X,Y,test_size=0.25)
```

```python
model= LinearRegression()
model.fit(x_train,y_train)
y_pred = model.predict(x_test)
```

```python
err = mse(y_test,y_pred)
print("mean sq err : ",err)
```
