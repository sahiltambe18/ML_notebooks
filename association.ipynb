```python
import csv
import pandas as pd
from mlxtend.preprocessing import TransactionEncoder
from mlxtend.frequent_patterns import apriori, association_rules

```

```python
dataset = []
with open("/content/Market_Basket_Optimisation - Market_Basket_Optimisation.csv") as file:
  reader = csv.reader(file, delimiter=",")
  for row in reader:
    dataset.append(row)
dataset
```

```python
# Assuming 'dataset' is defined and 'TransactionEncoder' and 'pd' are imported

te = TransactionEncoder()
df = te.fit_transform(dataset)
df = pd.DataFrame(df, columns=te.columns_)

# Find the column without a name
empty_col = df.columns[df.columns == ''].tolist()

# Drop the column without a name
if empty_col:
    df = df.drop(columns=empty_col[0])

df.head()
```

```python
item_set = apriori(df,min_support=0.01 , use_colnames=True)
item_set


```

```python
rules = association_rules(item_set, metric="confidence", min_threshold=0.20)
print(f" Confidence Level : 20%, No of rules : {len(rules)}")
rules
```

```python
rules = rules[["antecedents", "consequents", "support", "confidence"]]
rules
```
