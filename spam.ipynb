```python
df['spam'] = df['Category'].apply(lambda x:1 if x=='spam' else 0)
```

```python
df_new = df[['Message','spam']]
```

```python
from sklearn.model_selection import train_test_split as tts
```

```python
x_train , x_test , y_train , y_test = tts(df.Message,df.spam)
```

```python
from sklearn.feature_extraction.text import CountVectorizer

v = CountVectorizer()

xtrain_count = v.fit_transform(x_train.values)
xtrain_count
```

```python
from sklearn.naive_bayes import MultinomialNB
```

```python

model=MultinomialNB()
model.fit(xtrain_count,y_train)
```

```python
emails=["How are you brother?", "Free entry"]
email_count= v.transform(emails)
model.predict(email_count)
# print(email_count)
```

```python
x_test_count=v.transform(x_test)
model.score(x_test_count,y_test)
```
