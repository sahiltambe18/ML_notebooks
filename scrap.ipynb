```python
from bs4 import  BeautifulSoup
import requests
import pandas as pd
```

```python
Header = ("Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/123.0.0.0 Safari/537.36")
url = "https://www.amazon.in/electronics/b/?ie=UTF8&node=976419031&ref_=nav_cs_electronics"
```

```python
webpage = requests.get(url , Header)
```

```python
webpage.content
```

```python
soap = BeautifulSoup(webpage.content ,"html.parser" )
soap
```

```python
links = soap.find_all("a")
```

```python
link = links[4].get('href')
```

```python
new_url = "https://amazon.com"+link
```

```python
new_webpage = requests.get(new_url , Header)
```

```python
new_webpage
```

```python
new_soap = BeautifulSoup(new_webpage.content , "html.parser")
```
